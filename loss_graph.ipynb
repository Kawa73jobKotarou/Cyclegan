```python
# これを実行するとグラフゲット
import pandas as pd
import matplotlib.pyplot as plt
import os
import pdb

# path = "checkpoints/231114_miyazaki-ronbun_4fold_1_patchsize06/"
# path = "checkpoints/231110_miyazaki-ronbun_4fold_1_patchsize12/"
path = "checkpoints/2504015_allandpatch_0.5patchextention_3Dwide-ResNet_128patch_241004_allbody_fullset/"

# save_path=path+"img/"
save_path = '/home/kawanami/work/presen_images/250426/loss/2504015_allandpatch_0.5patchextention_3Dwide-ResNet_128patch_241004_allbody_fullset/'
if not os.path.exists(save_path):
    os.makedirs(save_path)
# テキストファイルからデータを読み込む
data = pd.read_csv(path+'final_loss_log.txt', sep=' ', header=None)
loss_name=['epoch', 'style', 'structure', 'perceptial', 'D_real', 'D_fake', 'G_GAN', "contrastive"]
num_columns, num_rows = data.shape
para = [1,0.1,0.4,0.1,1,1,0.4,1]
# para = [1,375,20,0.75,1,1,1,1]
print(data[1][14])
for k in range(len(data[1])):
    if data[1][k]=="inf":
        data[1][k]=data[1][k-1]
    print(str(k)+str(type(data[1][k])))
    
print(data[1][14])


# pdb.set_trace()
#各lossのパラメーターを考慮しないグラフを出力
for i in range(1,num_rows-1):
    plt.plot(data[0], data[i].astype(float) /para[i], label=loss_name[i]+'_loss')

    # グラフの設定
    plt.xlabel('Epoch')
    plt.ylabel('loss')
    plt.legend()
    # グラフを画像ファイルとして保存
    plt.savefig(save_path+loss_name[i]+'_loss.png')
    # グラフを表示
    plt.show()

# ジェネレータに関わるlossの出力
generator_all_loss=data[1].astype(float)+data[2]+data[3]+data[6]
plt.plot(data[0], data[1].astype(float), label=loss_name[1]+'_loss')
plt.plot(data[0], data[2], label=loss_name[2]+'_loss')
plt.plot(data[0], data[3], label=loss_name[3]+'_loss')
plt.plot(data[0], data[6], label=loss_name[6]+'_loss')
plt.plot(data[0], generator_all_loss, label='Generator_all_loss')
plt.ylim(0, 0.5)
# グラフの設定
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend()
# グラフを画像ファイルとして保存
plt.savefig(save_path+'Generator_all_loss.png')
# グラフを表示
plt.show()

# ディスクリミネータに関わるlossの出力
D_loss=(data[4]+data[5])/2
plt.plot(data[0], data[4], label=loss_name[4]+'_loss')
plt.plot(data[0], data[5], label=loss_name[5]+'_loss')
plt.plot(data[0], D_loss, label='Discriminator_all_loss')
plt.ylim(0, 0.5)

# グラフの設定
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend()
# グラフを画像ファイルとして保存
plt.savefig(save_path+'D_loss.png')
# グラフを表示
plt.show()
```
